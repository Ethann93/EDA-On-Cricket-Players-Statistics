### Split the span into start and end date


```python
df['Span'].str.split(pat= '-')
```

```python
df['Debut_Year']= df['Span'].str.split(pat = '-').str[0]
```

```python
df['Final_Year']= df['Span'].str.split(pat = '-').str[1]
```

```python
df
```

### Drop Span Columns

```python
df = df.drop(['Span'], axis =1)
```

### Split the country from the player

```python
df['Player'].str.split(pat = '(')
```

```python
df['Country'] = df['Player'].str.split(pat = '(').str[1]
```

```python
df['Country'] = df['Country'].str.split(pat = ')').str[0]
```

```python
df['Country'] 
```

```python
df['Player'] = df['Player'].str.split(pat = '(').str[0]
```

### Create a career length column

```python
df['Career_length'] = (df['Final_Year'] - df['Debut_Year'])
```

# Questions

### 1. What is the average career length?

```python
df['Career_length'].mean()
```

The average career length is 12.75 years

### 2. What is the average Batting_strike_rate for cricketers who played over 10 years?

```python
df[df['Career_length'] > 10]['Batting_Strike_Rate'].mean()
```

### 3. Find number of cricketers who played before 1960 

```python
df[df['Debut_Year'] <1960]['Player'].count()
```

### 4. Max highest Inns score by country

```python
df.groupby('Country')['Highest_Inns_Score'].max().to_frame('Highest In Country').reset_index().sort_values('Highest In Country', ascending = False)
```

### 5. Hundreds, Fifties, ducks (0) AVG by country

```python
df.groupby('Country')[['100.0','50.0','0.0']].mean()
```
